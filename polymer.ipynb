```python
import numpy as np
import scipy as sp
import random
from itertools import count
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from numba import prange,jit
```

```python
%matplotlib notebook
plt.style.use('fivethirtyeight')

box = 20

pos = np.zeros([1,2])
phi = np.zeros([1])
r   = 0.2
mks = 3

@jit(fastmath=True)
def rand_update(position):
    position[0,0] += (-1)**random.randint(0,1)*random.random()
    position[0,1] += (-1)**random.randint(0,1)*random.random()

def angle_update(angle):
    angle[0] = angle[0] + (-1)**random.randint(0,1)*random.random()*np.pi/10

@jit(fastmath=True)
def animate(i):
    rand_update(pos)
    angle_update(phi)
    plt.clf()
    plt.axis(( -1*box, box, -1*box, box))
    plt.plot(pos[0,0], pos[0,1], 'ro', markersize=mks)
    plt.plot(pos[0,0] + r*np.cos(phi), pos[0,1] + r*np.sin(phi), 'bo', markersize=mks)

ani = FuncAnimation(plt.gcf(), animate, interval=500)
plt.tight_layout()
plt.show()
```

```python
%matplotlib notebook
plt.style.use('fivethirtyeight')

box = 50 / 2

pos = np.zeros([1,2])
phi = np.zeros([1])
x = y = np.zeros(2)
r   = 1.0


@jit(fastmath=True)
def rand_update(position):
    position[0,0] += (-1)**random.randint(0,1)*random.random()*0.5
    position[0,1] += (-1)**random.randint(0,1)*random.random()*0.5

def angle_update(angle):
    angle[0] = angle[0] + (-1)**random.randint(0,1)*random.random()*np.pi/5
    
@jit(fastmath=True)
def animate(i):
    rand_update(pos)
    angle_update(phi)
    x = [pos[0,0], pos[0,0] + r*np.cos(phi)]
    y = [pos[0,1], pos[0,1] + r*np.sin(phi)]
    plt.clf()
    plt.axis(( -1*box, box, -1*box, box))
    plt.plot(x[0], y[0], 'ro', markersize=2)
    plt.plot(x[1], y[1], 'bo', markersize=2)

ani = FuncAnimation(plt.gcf(), animate, interval=1000)
plt.tight_layout()
plt.show()
```
